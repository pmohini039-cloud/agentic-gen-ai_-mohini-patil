```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('darkgrid')

f, axes = plt.subplots(2, 2, figsize=(15, 15))

# Top-left
k1 = sns.kdeplot(
    x=movies.BudgetMillions,
    y=movies.AudienceRating,
    ax=axes[0, 0]
)

# Top-right
k2 = sns.kdeplot(
    x=movies.BudgetMillions,
    y=movies.CriticRating,
    ax=axes[0, 1]
)

# Set limits
k1.set(xlim=(-20, 160))
k2.set(xlim=(-20, 160))

# Bottom-left (violin plot)
z = sns.violinplot(
    data=movies[movies.Genre == 'Drama'],
    x='Year',
    y='CriticRating',
    ax=axes[1, 0]
)

# Bottom-right (KDE with fill)
k4 = sns.kdeplot(
    x=movies.CriticRating,
    y=movies.AudienceRating,
    fill=True,
    cmap='Reds',
    ax=axes[1, 1]
)

# Optional contour lines on top
k4b = sns.kdeplot(
    x=movies.CriticRating,
    y=movies.AudienceRating,
    cmap='Reds',
    ax=axes[1, 1]
)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Dark theme with black background
sns.set_style('dark', {'axes.facecolor': 'black'})

f, axes = plt.subplots(2, 2, figsize=(15, 15))

# 🔹 Plot [0,0]
k1 = sns.kdeplot(
    x=movies.BudgetMillions,
    y=movies.AudienceRating,
    fill=True,
    cmap='inferno',
    ax=axes[0, 0]
)

k1b = sns.kdeplot(
    x=movies.BudgetMillions,
    y=movies.AudienceRating,
    cmap='cool',
    ax=axes[0, 0]
)

# 🔹 Plot [0,1]
k2 = sns.kdeplot(
    x=movies.BudgetMillions,
    y=movies.CriticRating,
    fill=True,
    cmap='inferno',
    ax=axes[0, 1]
)

k2b = sns.kdeplot(
    x=movies.BudgetMillions,
    y=movies.CriticRating,
    cmap='cool',
    ax=axes[0, 1]
)

# 🔹 Plot [1,0]
z = sns.violinplot(
    data=movies[movies.Genre == 'Drama'],
    x='Year',
    y='CriticRating',
    palette='coolwarm',
    ax=axes[1, 0]
)

# 🔹 Plot [1,1]
k4 = sns.kdeplot(
    x=movies.CriticRating,
    y=movies.AudienceRating,
    fill=True,
    cmap='Blues_r',
    ax=axes[1, 1]
)

k4b = sns.kdeplot(
    x=movies.CriticRating,
    y=movies.AudienceRating,
    cmap='gist_gray_r',
    ax=axes[1, 1]
)

# Axis limits
k1.set(xlim=(-20, 160))
k2.set(xlim=(-20, 160))

# Titles (important for dashboards)
axes[0,0].set_title("Budget vs Audience (Inferno + Cool)", color='white')
axes[0,1].set_title("Budget vs Critic (Inferno + Cool)", color='white')
axes[1,0].set_title("Drama Critic Ratings by Year", color='white')
axes[1,1].set_title("Critic vs Audience Density", color='white')

plt.tight_layout()
plt.show()
```

## 🔍 Final Insights & Conclusion
