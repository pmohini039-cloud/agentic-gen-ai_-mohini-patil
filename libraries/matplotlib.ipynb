```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(0,10,50)

plt.plot(x1, np.sin(x1), '-')
plt.plot(x1, np.cos(x1), '-')
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x1,np.cos(x1),'*')
```

```python
#create plot figure
plt.figure()   
#create the first of two panels and set current axis
plt.subplot(2,1,1)    #(rows,columns and panel number)
plt.plot(x1,np.sin(x1))
#create the second of two panels and set current axis
plt.subplot(2,1,2)    #(rows,columns and panel number)
plt.plot(x1,np.cos(x1))
```

```python
x=np.linspace(0,2,100)                 #state-machine interface
plt.plot(x,x,label='linear')
plt.plot(x,x**2,label='quadratic')
plt.plot(x,x**3,label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("simple plot")
plt.legend()
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro')    #formatting the style of plot
plt.axis([0,6,0,20])
plt.show()
```

```python
#evently sampled time at 200ms intervals
t=np.arange(0.,5.,0.2)   
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^')   #red dashes blue square and green traiangles
plt.show()
```

```python
#sine and cosine curves using object oriented API
fig,ax=plt.subplots(2)  #first create a grid of plots/ ax will be an array of two axes objects
ax[0].plot(x1,np.sin(x1),'b--')   #call plot() method on an appropriate object
ax[1].plot(x1,np.cos(x1),'b--');
```

```python
fig=plt.figure()                        #objects and references
x2=np.linspace(0,5,10)
y2=x2**2
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x2,y2,'r')
axes.set_xlabel('x2')
axes.set_ylabel('y2')
axes.set_title('title')
```

```python
data2=[5.,25.,50.,20.]                #bar chart
plt.bar(range(len(data2)),data2)
plt.show()
```

```python
data2=[5.,25.,50.,20.]                # horizontal bar chart
plt.barh(range(len(data2)),data2)
plt.show()
```

```python
x9=np.arange(0,4,0.2)                 #error bar
y9=np.exp(-x9)
e1=0.1*np.abs(np.random.randn(len(y9)))
plt.errorbar(x9,y9,yerr=e1,fmt='.-')
plt.show()
             
```

```python
A=[15.,30.,45.,22.]            #stacked bar chart
B=[15.,25.,50.,20.]
z2=range(4)
plt.bar(z2,A,color="b")
plt.bar(z2,B,color='r',bottom=A)
plt.show()
```

```python
plt.figure(figsize=(7,7))               #pie chart
x10=[35,25,20,20]
labels=['Computers','Electronics','Mechanical','Chemical']
plt.pie(x10,labels=labels)
plt.show()
```
