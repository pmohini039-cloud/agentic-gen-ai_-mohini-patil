```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()

data = iris.data[iris.target == 0]

df = pd.DataFrame(data, columns=iris.feature_names)

sns.kdeplot(
    x=df['sepal length (cm)'],
    y=df['sepal width (cm)'],
    cmap='viridis',
    fill=True
)

plt.xlabel('Sepal Length Cm')
plt.ylabel('Sepal Width Cm')
plt.title('Iris-setosa')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

# Load data
iris = load_iris()

# Convert to DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['Species'] = iris.target
df['Species'] = df['Species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

sns.set_style('darkgrid')

f, axes = plt.subplots(2, 2, figsize=(15, 15))

sns.boxplot(x="Species", y="petal length (cm)", data=df, ax=axes[0,0])
sns.violinplot(x='Species', y='petal length (cm)', data=df, ax=axes[0,1])
sns.stripplot(
    x='Species',
    y='sepal length (cm)',
    data=df,
    jitter=True,
    edgecolor='gray',
    size=8,
    palette='winter',
    orient='v',
    ax=axes[1,0]
)

axes[1,1].hist(df['petal length (cm)'], bins=100)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load dataset
iris = load_iris()

# Convert to DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add Species column
df['Species'] = iris.target

# Convert to names + category
df['Species'] = df['Species'].map({
    0: 'setosa',
    1: 'versicolor',
    2: 'virginica'
}).astype('category')

# Now your logic
list1 = []
mylabels = []

for gen in df['Species'].cat.categories:
    list1.append(df[df['Species'] == gen]['sepal length (cm)'])
    mylabels.append(gen)

plt.hist(list1, bins=30, stacked=True, rwidth=1, label=mylabels)
plt.legend()
plt.show()
```
