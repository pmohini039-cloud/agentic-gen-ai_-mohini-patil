```python
print(bin(20))          #Integer to binary
print(int(0b10100))     #Binary to integer
```

```python
print(oct(25))      #Integer to octal
print(int(0o31))    #Octal to integer
```

```python
print(hex(20))      #Integer to hexadecimal
print(int(0x14))    #hexadecimal to integer
```

Bitwise Operator

```python
print(~1)      #Complement
print(~52)
```

```python
print(12 & 15)    #AND Operator
print(26 & 59)
print(49 & 16)
```

```python
print(15 | 14)    #OR Operator
print(12 | 13)
```

```python
print(12 ^ 13)   #EXCLUSIVE OR
print(12 ^ 14)
print(13 ^ 12)
```

```python
print(12<<1)     #LEFT SHIFT
print(12<<2)
```

```python
print(12>>1)     #RIGHT SHIFT
print(12>>2)
```
