copy_set

```python
myset={'one','two','three','four','five','six','seven','eight','nine','ten'}
myset
```

```python
myset1=myset    #create a reference myset1
myset1 
```

```python
id(myset1), id(myset)    #the address of both the set will be same
```

```python
my_set=myset.copy()   #create copy of the list
my_set
```

```python
id(my_set) #the address of my_set will be different from myset
```

```python
myset.add('eleven')
myset
```

```python
myset1   #myset1 will be also impacted as it is pointing to the same set
```

```python
my_set    #copy of the set won't be impacted due to changes made on the original set
```

set_operation

Union

```python
A={1,2,3,4,5}
B={4,5,6,7,8}
C={8,9,10}
```

```python
A|B   #A union B means all the elements from set A and set B/ no duplication is allowed
```

```python
A.union(B)
```

```python
A.union(B,C)  #union of A,B and C
```

```python
A.update(B,C)   #updates the set calling the update() method with union of A,B and C
A
```
