```python
#Right angled triangle pattern
for i in range(1,6):
    print(' * ' *i)
```

```python
#inverted right angled triangle pattern
for i in range (5,0,-1):
    print(' * ' *i)
```

```python
# Pyramid pattern
n = 5  
for i in range(1, n+1):
    print(' ' * (n - i) + '* ' * (2*i - 1))
```

```python
#inverted pyramid pattern
n=5
for i in range (n,0,-1):
    print(' '* (n-i) + '* ' * (2*i - 1))
```

```python
# Diamond pattern
n = 5  
for i in range(1, n):
    print(' ' * (n - i) + '* ' * i)
for i in range(n, 0, -1):
    print(' ' * (n - i) + '* ' * i)
```

```python
#hallow square shape
for i in range (5):
    for j in range (5):
        if i==0 or i==4 or j==0 or j==4:
            print('*',end='')
        else:
            print(' ',end='')
    print()
```

```python
#full square pattern
for i in range (5):
    print(' * ' *5)
```

```python
#right angled triangle number pattern
for i in range (1,6):
    print(' '.join(str(x) for x in range(1, i+1)))
```

```python
#invertd right angled triangle number pattern
for i in range (5,0,-1):
    print(' '.join(str(x) for x in range(1,i+1)))
```

```python
#floyd's triangle
num=1
for i in range (1,6):
    for j in range (1,i+1):
        print(num,end=' ')
        num += 1
    print()
```

```python
#hallow right angled triangle
for i in range (1,6):
    for j in range (1,i+1):
        if j==1 or j==i or i==5:
            print('*',end=' ')
        else:
            print(' ',end=' ')
    print()
```

```python
#hallow pyramid pattern
for i in range (1,6):
    for j in range (5 - i):
        print(' ',end=' ')
    for j in range (2 * i - 1):
        if j==0 or j==2*i-2 or i==5:
            print('*',end=' ')
        else:
            print(' ',end=' ')
    print()
```

```python
#hallow diamond pattern

n = 5
for i in range(1, n + 1):
    for j in range(n - i):
        print(' ', end=' ')
    for j in range(2*i - 1):
        if j == 0 or j == 2*i - 2:
            print('*', end=' ')
        else:
            print(' ', end=' ')
    print()


for i in range(n - 1, 0, -1):
    for j in range(n - i):
        print(' ', end=' ')
    for j in range(2*i - 1):
        if j == 0 or j == 2*i - 2:
            print('*', end=' ')
        else:
            print(' ', end=' ')
    print()
    
```

```python
#butterfly number pattern
n = 5
for i in range(1, n+1):
    for j in range(1, i+1):
        print(j, end=' ')
    for j in range(2*(n-i)):
        print('  ', end='')
    for j in range(i, 0, -1):
        print(j, end=' ')
    print()

for i in range(n, 0, -1):
    for j in range(1, i+1):
        print(j, end=' ')
    for j in range(2*(n-i)):
        print('  ', end='')
    for j in range(i, 0, -1):
        print(j, end=' ')
    print()
```

```python
#butterfly star pattern
n = 5
for i in range(1, n+1):
    for j in range(i):
        print('*', end=' ')
    for j in range(2*(n-i)):
        print(' ', end=' ')
    for j in range(i):
        print('*', end=' ')
    print()

for i in range(n, 0, -1):
    for j in range(i):
        print('*', end=' ')
    for j in range(2*(n-i)):
        print(' ', end=' ')
    for j in range(i):
        print('*', end=' ')
    print()
```

```python
#hallow number pyramid
n=5
for i in range (1,n+1):
    for j in range (n-i):
        print(' ',end=' ')

    for j in range (1,2*i):
        if j==1 or j==2*i-1 or i==n:
            print(i,end=' ')
        else:
            print(' ',end=' ')
    print()
```

```python
#full star pyramid
n=5
for i in range (1,n+1):
    for j in range (n-i):
        print(' ',end=' ')

    for j in range (2*i-1):
        print('*',end=' ')
    print()
```

```python
#inverted full star pyramid
n=5
for i in range (n, 0, -1):
    for j in range (n-i):
        print(' ',end=' ')

    for j in range (2*i-1):
        print('*',end=' ')
    print()
```
