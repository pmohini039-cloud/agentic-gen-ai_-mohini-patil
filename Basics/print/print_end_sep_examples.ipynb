Using .format()

```python
num1=20
num2=40
add=num1+num2
print('The addition of {} and {} is {}'.format(num1,num2,add))
```

Using f strings

```python
name='Mohini'
age=21
city='jalgaon'
print(f'Hello my name is {name} and i am {age} years old  from {city}')  #f means formatted string and It allows you to embed variables and expressions inside {}
```
