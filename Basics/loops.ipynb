loops

while loop

```python
i=1
while i <= 5:
    print("data science")
    i=i+1
```

```python
i=1
while i <= 5:
    print('data science',':',i)
    i=i+1
```

nested while loop

```python
i=1
while i <= 5:
    print('data science')
    j=1
    while j <= 4:
        print('technology')
        j=j+1

    i=i+1
    print()
```

```python
i=1
while i <= 5:
    print('data science', end=' *^ ')   #when we mention end then the new line will not be created
    j=1
    while j <= 4:
        print('technology',end=' *^ ')
        j=j+1
    i=i+1
    print()
```

```python
i=1
while i <= 2:
    j=0
    while j <= 2:
        print(i*j,end=' ')
        j+=1
    print()
    i+=1
```

```python
i=1
while i <= 4:
    j=0
    while j <= 3:
        print(i*j,end=' ')
        j+=1
    print()
    i+=1
```

for loop

```python
for i in range(5):
    print(i)
```

```python
for i in range(1,10,3):
    print(i)
```

```python
for i in range(1,51):     #print the number which is divisible by 5
    if i % 5==0:
        print(i)
```

```python
for i in range(1,11):
    print(i)
```

```python
for i in range (1,11):
    if i==3:
        break    #the break is use to immediately stop the loop
    print(i)
```

```python
for i in range (1,11):
    if i==3:
        continue    #the continue will skip the one step and continues the loop
    print(i)
```

```python
for i in range(1,11):
    pass    #pass does nothing but avoids an error
```

```python
#write a code in which the user ask for a chocolate from vendor machine
x=int(input("how many chocolate you want:?"))
i=1
while i <= x:
    print("chocolate")
    i+=1
```

```python
#machine has only 5 chocolates
available_chocolates = 5
x = int(input("how many chocolates you want:?"))
i = 1

while i <= x:
    if i > available_chocolates:
        print("Out of stock, bye for now")
        break
    print("chocolate")
    i += 1
```
