if you wab=nt to call any function or package you have to use keywords like import or from

```python
import math as m
print(m.sqrt(25))
print(round(m.sqrt(25)))
print(m.floor(6.9))
print(m.ceil(6.9))
print(m.pow(2,3))
```

```python
from math import sqrt, ceil, floor, pow
print(sqrt(25))
print(floor(6.9))
print(ceil(6.9))
print(pow(2,3))
```
