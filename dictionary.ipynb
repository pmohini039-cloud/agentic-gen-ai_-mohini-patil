copy dictionary

```python
mydict = {'Name':'Mohini' , 'ID': 74123 , 'DOB': 2004 , 'job' :'Analyst'} 
mydict
```

```python
mydict1=mydict   #create a new reference
```

```python
id(mydict),id(mydict1)     #the address of  both dictionary will be same
```

```python
mydict2=mydict.copy()    #create a copy of the dictionary
```

```python
id(mydict2)    #the address of the mydict2 will be different from  mydict
```

```python
mydict['Address']='Hyderabad'
mydict
```

```python
mydict1     #mydict1 will also imapcted as is pointing to the same dictionary
```

```python
mydict2     #copy of list won't be impacted due to the changes made in the original list
```
